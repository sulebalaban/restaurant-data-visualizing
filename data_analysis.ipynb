```python
import pandas as pd 
import numpy as np 
import seaborn as sns 
import matplotlib.pyplot as plt 
```

```python
df = pd.read_csv('Zomato data .csv')


```

```python
df['rate']=df['rate'].str.replace('/5','')
```

```python
df['rate']=df['rate'].astype(float)
```

```python
df.info()
```

```python
sns.countplot(x=df['listed_in(type)'],data=df,hue='listed_in(type)',palette='pastel',legend=False)
plt.xlabel('Type of Restaurant')
```

```python
group_data=df.groupby('listed_in(type)')['votes'].sum()
result=pd.DataFrame({'votes':group_data})
plt.plot(result,c='blue',marker='o')
plt.xlabel('Type of restaurant',size= 20)
plt.ylabel('Votes',size=20)
```

```python
sns.countplot(x='online_order', data=df, hue='online_order', palette='Set2', legend=False)


```

```python
couple_data=df['approx_cost(for two people)']
sns.countplot(x=couple_data, hue=couple_data, palette='pastel', legend=False)
plt.xlabel('Cost for two people',size=15)
```
